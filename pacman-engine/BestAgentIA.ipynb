```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.preprocessing import LabelEncoder
df=pd.read_csv("./agentsData.cvs",names=["pacman_pos_x","pacman_pos_y","ghost1_pos_x","ghost1_pos_y","distance_ghost_1","ghost2_pos_x","ghost2_pos_y","distance_ghost_2","distance_ghost_scared","distance_ghost_scared_2","closest_food1_x","closest_food1_y","capsule1_pos_x","capsule1_pos_y","distance_capsule_1","capsule2_pos_x","capsule2_pos_y","distance_capsule_2","score","direction"])
```

```python
print(df.head())
```

```python
#Enconding directions into numbers:
label_encoder = LabelEncoder()
df["direction_encoded"] = label_encoder.fit_transform(df["direction"])
```

```python
print(df.head(100))
```

```python
#selected_columns = df[["direction", "direction_encoded"]]
#selected_rows = df[df['direction_encoded'] == 3]
unique_values = df['direction_encoded'].unique()
unique_values
```

```python
df = df.drop("direction", axis=1)
df
```

```python
y_train = df.iloc[:, -1].astype(np.float32)
x_train = df.iloc[:, :-1].astype(np.float32)
```

```python
dataset_size=len(x_train)
train_set_size=int(dataset_size*0.7)
x_val,y_val=x_train[train_set_size:],y_train[train_set_size:]
x_train,y_train=x_train[:train_set_size],y_train[:train_set_size]
```

```python
num_inputs=x_train.shape[1] # this works out how many columns there are in x, i.e. how many inputs our network needs.
num_outputs=5
```

```python
model = keras.Sequential(name="best_agent")
layer1=layers.Dense(100, activation="relu", input_shape=(num_inputs,))
model.add(layer1)
layer2=layers.Dense(100, activation="relu")
model.add(layer2)
layer3=layers.Dense(100, activation="relu")
model.add(layer3)
layer4=layers.Dense(50, activation="relu")
model.add(layer4)
layer5=layers.Dense(10, activation="relu")
model.add(layer5)
final_layer=layers.Dense(num_outputs, activation="softmax")
model.add(final_layer)
```

```python
model.compile(
    optimizer=keras.optimizers.Adam(0.001),  # Optimizer
    # Loss function to minimize
    loss=keras.losses.SparseCategoricalCrossentropy(),
    metrics='accuracy'
)
```

```python
# Do the usual business for keras training
history = model.fit(
    x_train,
    y_train,
    batch_size=250,
    epochs=2000,
    validation_data=(x_val, y_val),verbose=1
)
```

```python
model.save('best_agent.h5',save_format='h5')
```
